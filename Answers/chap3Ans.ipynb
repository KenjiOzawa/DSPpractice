## 3.3 スペクトル

(1-1)  
\begin{eqnarray}
 p (t) = 2 \cos (2 \pi f_0 t) - \sin (2 \pi f_0 t) 
\end{eqnarray}

【答】
+ $A_n$ と $B_n$ を読み取る。基本波 $n=1$だけが含まれています。
    + 基本波：$A_1 = 2$, $B_1 = -1$
+ それゆえ 
    + $C_1 = \frac{1}{2}(2 + j1)$, $C_{-1} = \frac{1}{2}(2 - j1)$

(1-2)  
\begin{eqnarray}
  p (t) = 2 \sqrt{5} \cos \left( 2 \pi f_0 t - \tan^{-1} \frac{1}{2} \right) + \sqrt{5} \cos \left( 2 \pi (2 f_0) t + \tan^{-1} 2 \right) 
\end{eqnarray}

+ $|D_n|$ と $\phi_n$ を読み取る。基本波 $n=1$ と第2高調波 $n=2$ が含まれている。
    + 基本波：$|D_1| = 2 \sqrt{5}$, $\phi_1 = - \tan^{-1} \frac{1}{2}$
    + 第2高調波：$|D_2| = \sqrt{5}$, $\phi_2 =  \tan^{-1} 2$
+ それゆえ 
    + $C_1 = \frac{1}{2}(2 \sqrt{5}) \exp(-j \tan^{-1} \frac{1}{2}) = \sqrt{5} \exp(-j \tan^{-1} \frac{1}{2})$, $C_{-1} =\sqrt{5} \exp(j \tan^{-1} \frac{1}{2})$　
    + $C_2 = \frac{1}{2}\sqrt{5} \exp(j \tan^{-1} 2)$, $C_{-2} =\frac{\sqrt{5}}{2} \exp(-j \tan^{-1} 2)$ 
    
波形とスペクトルは以下のセルをコピー・ペーストして実行すれば表示されます。

````python
# === 波形を描画します ===
fs = 44100.0    # 標本化周波数は 44.1 kHz とします。
f0 = 500.0; 
D1 = 2 * np.sqrt(5.0) ; phy1 = -np.arctan(1.0/2.0)
D2 = np.sqrt(5.0) ;     phy2 = -np.arctan(2.0)

t = np.arange(0, 1, 1/fs)     # 横軸の範囲 1 s 間を，1 / fs [s] ごとに区切る時刻を配列 t に納めます。，
pt = D1 * np.cos( 2.0 * np.pi * f0 * t + phy1) + D2 * np.cos( 2.0 * np.pi * (2*f0) * t + phy2)
    
plt.subplot(3, 1, 1) # 画面を3行1列に分けて，その1番上を選択
plot_wave(t[0:300], pt[0:300])

# === 複素スペクトルをセットする関数を定義します ===
def set_cmplxC(Cn_half):
    Cn_plus = Cn_half[1: ]              # 正の周波数については，与えられた C_n を埋めます。
    Cn_minus = np.conj(Cn_plus[ : :-1]) # 負の周波数については，配列（リスト）を逆順に納め，複素共役をとります。
    C0 = Cn_half[0]
    Cn = np.r_[Cn_minus, C0, Cn_plus]   # 負の周波数成分，直流成分，正の周波数成分の C_n を並べます。
    return (Cn)

# === スペクトルを描画します ===
n_max = 10
Cn_half = np.zeros(n_max, dtype='complex') # 0を詰めた配列を準備します。
Cn_half[1] = np.sqrt(5.0)     * np.exp(-1j * np.arctan( 1/2.0 ))
Cn_half[2] = np.sqrt(5.0)/2.0 * np.exp(1j * np.arctan( 2.0 ))
Cn = set_cmplxC(Cn_half)
draw_cmplx_spectrum(f0, Cn)
````

(1-3)  
\begin{eqnarray}
p (t) = \frac{1}{\pi} + \frac{1}{2} \sin (2 \pi f_0 t) - \frac{2}{\pi} \left\{ \frac{1}{3} \cos(2 \pi (2 f_0) t ) + \frac{1}{15} \cos(2 \pi (4 f_0) t ) + \cdots +\frac{1}{(4k^2-1)} \cos( 2 \pi (2k f_0) t )+ \cdots  \right\}
\end{eqnarray}

【答】
+ $A_n$ と $B_n$ を読み取る。直流成分と基本波と偶数次の高調波が含まれている。
    + 直流成分：$A_0 = \frac{1}{\pi}$
    + 基本波：$A_1 = 0$, $B_1 = \frac{1}{2}$
    + 偶数次高調波： $A_{2k} = -\frac{2}{\pi}\frac{1}{4 k^2 -1}= -\frac{2}{\pi} \frac{1}{(2 k)^2 -1}$, $B_{2k} = 0$  
    + 奇数次高調波： $A_{2k-1} = 0$, $B_{2k-1} = 0$ 
+ それゆえ 
    + $C_0 = \frac{1}{\pi}$
    + $C_1 = \frac{1}{2}(0 - j\frac{1}{2})=-j\frac{1}{4}$, $C_{-1} = j\frac{1}{4}$
    + $C_{\pm 2k} = -\frac{1}{\pi}\frac{1}{4 k^2 -1}$, $C_{\pm (2k-1)} = 0$

波形とスペクトルは以下のセルをコピー・ペーストして実行すれば表示されます。波形は「正弦波を半波整流した結果」です。

````python
# === 第n高調波の波形を生成する関数を定義します。 ===
def the_Nth_wave(n, f0, t):
    '''第n高調波の波形の定義'''
    return( (-2.0/np.pi)*(1.0/(n**2-1)) * np.cos(2 * np.pi * (n * f0) * t) )

# === 波形を描画します ===
fs = 44100.0    # 標本化周波数は 44.1 kHz とします。
f0 = 500.0; 

t = np.arange(0, 1, 1/fs)     # 横軸の範囲 1 s 間を，1 / fs [s] ごとに区切る時刻を配列 t に納めます。，

n_start = 2 # 足し込む最低の高調波番号
n_end = 21  # 足し込む最高の高調波番号
n_skip = 2  # 高調波番号のスキップ幅（この場合は偶数のみ）

A0 = 1.0/np.pi; B1 = 1.0/2.0 
sum = A0 + B1 * np.sin(2 * np.pi * f0 *t) # 直流成分と基本波成分をセットします。
for n in range(n_start, n_end, n_skip):               # 偶数次高調波を足していきます。
    component = the_Nth_wave(n, f0, t)    # 成分音として第n高調波を計算
    sum += component                     # 成分音を足しこむ
    
plt.subplot(3, 1, 1) # 画面を3行1列に分けて，その1番上を選択
plot_wave(t[0:300], sum[0:300], color = 'b')

# === スペクトルを描画します ===
n_max = n_end
Cn_half = np.zeros(n_max, dtype='complex') # 0を詰めた配列を準備します。
Cn_half[0] = A0
Cn_half[1] = -1j * 1.0/4.0
for n in range(n_start, n_end, n_skip):               # 偶数次高調波を足していきます。
    Cn_half[n] = (-1.0/np.pi)*(1.0/(n**2-1))
Cn = set_cmplxC(Cn_half)
draw_cmplx_spectrum(f0, Cn)
````
