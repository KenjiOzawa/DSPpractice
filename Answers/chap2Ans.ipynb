## 2.2 正弦波を生成して聞いてみる
(1) 周波数: 500 Hz，振幅: 1, 継続時間: 2 s の余弦波を生成して聞いてみなさい。ただし，標本化周波数: 44.1 kHz とする。

```python
import numpy as np
from IPython.display import Audio

f  = 500.0      # 500 Hz の音を合成します。
A  = 1.0        # 振幅は 1 とします。
fs = 44100.0    # 標本化周波数は 44.1 kHz とします。

t = np.arange(0, 2, 1/fs)     # 横軸の範囲 1 s 間を，1 / fs [s] ごとに区切る時刻を配列 t に納めます。，
sampledWave = A * np.cos(2.0 * np.pi * f * t)  # 標本値を配列 sampledWave に納めます。

audio = Audio(sampledWave, rate = fs)  # 引数 rate に標本化周波数を指定します。
audio                                  # オーディオオブジェクト audio を再生します。
```

```python
f1  = 500.0     #  500 Hz の音を合成します。
f2  = 1000.0    # 1000 Hz の音を合成します。
A1  = 1.0       #  500 Hz 成分の振幅は 1 とします。
A2  = 0.5       # 1000 Hz 成分の振幅は 0.5 とします。
fs = 44100.0    # 標本化周波数は 44.1 kHz とします。

t = np.arange(0, 2, 1/fs)     # 横軸の範囲 1 s 間を，1 / fs [s] ごとに区切る時刻を配列 t に納めます。，
sampledWave = A1 * np.cos(2.0 * np.pi * f1 * t)  + A2 * np.cos(2.0 * np.pi * f2 * t) # 標本値を配列 sampledWave に納めます。

audio = Audio(sampledWave, rate = fs)  # 引数 rate に標本化周波数を指定します。
audio                                  # オーディオオブジェクト audio を再生します。
```

(2) CD音質のディジタルデータ（標本化周波数: 44.1 kHz, 量子化精度: 16 bit, チャネル数: 2 ch）について，ビットレート（bps = {bit per second} = 1 s あたりのビット数）を求めなさい。

```python
fs = 44100 # 標本化周波数 [sample/s]
qb = 16    # 量子化ビット数 [bit/sample]
ch = 2     # チャネル数 [ ]
bit_rate = fs * qb * ch
print('bit rate =', bit_rate, '[bps]')
```

(3) 上記の量子化精度 4 bit データについて，ビットレートを求めなさい。（→【答】以下に計算したとおりの値になります。MP3 では 128 Kbps 以上のビットレートであれば十分な音質が得られますので，知覚符号化の効果は大きいことが分かります。）

```python
qb = 4    # 量子化ビット数 [bit/sample]
bit_rate = fs * qb * ch
print('bit rate =', bit_rate/1024, '[Kbps]')
```

## 2.4 音圧の dB 表示
(1) 音圧レベル 114 dB の音の音圧は何 Pa か答えなさい（ヒント：20 $\mu$Pa より $120 - 6$ [dB] だけレベルが高い）。

【答】 20 dB アップする度に音圧は 10 倍されます。それゆえ，40 dB アップで 100 倍，60 dB アップで 1000倍，... 120 dB アップで 1000000倍（＝10$^6$倍）されます。$\mu$ は 10$^{-6}$ ですから，120 dB アップで 20 Pa となります。$-$6 dB で音圧は 1/2 倍です。それゆえ，114 dB の音圧は 10 Pa です。 

(2) 部屋の響きの程度を表す指標である**残響時間**は， 「室内に放射された音が平衡状態に達したのち，音を急に停止し，その後の残響のエネルギー密度が音源停止直前のエネルギー密度に比べて $-60$ dB になるまでの時間を秒単位で表したもの」である。さて， $-60$ dB とは，エネルギー密度の比として「何分の1」であるか答えなさい。

【答】 10 dB ダウンする度にパワーは 1/10 倍されます。それゆえ，20 dB ダウンで 1/100 倍, 30 dB ダウンで 1/1000倍，... 60 dB ダウンで 1/1000000 倍となります。答えは「100万分1」です。
