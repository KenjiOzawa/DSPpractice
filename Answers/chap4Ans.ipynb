## 4.1 ディジタル信号のフーリエ変換

(1) 上記 (b) の標本化について，f0 = 16.0 として標本化された波形を観察しなさい。一見すると，正しく標本化されているようである。しかし，標本化周波数 $f_s$ が 32 Hz の場合には，ナイキスト周波数である 16 Hz の波は正しく標本化できないはずである。このことを実感するために，周波数 16 Hz の "sin 波" を検討対象とするようにプログラムを書き換えて，どのような点が標本化されるか調べてみること。

【答】以下のとおり，# *** コメント *** を残した3行を修正してください。→実行すると「サンプルされた値は，全て 0 」となります。これでは，元の sin波の振幅は，全く分かりません。パワースペクトルの縦軸は 1e-14 という値が見られるとおり，「ゴミ」です。

```python
fs = 32.0
f0 = 16.0     # ***** 16 Hz の波を標本化します。 *****
A  = 1.0

t = np.arange(0, 1, 1/fs)
pn = A * np.sin(2.0 * np.pi * f0 * t)  # ***** sin波の標本値を配列 fn に納めます。*****
sp = np.fft.fft(fn)

plt.subplot(3,1,1)
# 参考とするため，疑似的なアナログ波形を描く
fs_a = 1000.0
t_a = np.arange(0, 1, 1/fs_a)
pn_a = A * np.sin(2.0 * np.pi * f0 * t_a)  # ***** sin波の標本値を配列 fn_a に納めます。 *****
# ディジタル波形を描く
plot_wave(t_a, pn_a, hold = True) 
plot_wave(t, pn, marker = 'o')

plt.subplot(3,1,2)
draw_FFT_spectrum(sp, fs)
```

(2) 検討対象とする波を sin 波に替えた場合について，f0 の値を 2, 3, 4, ..., 15 Hz と変えながら，波形とスペクトルを描きなおすこと（上限は 15 Hz までです！）。そのとき，「納得できること」と「納得できないこと」は同様に観察できるか，確認しなさい。

【答】同様に観察できるはずです。すなわち，
(A) 納得できること：「(A-1) 振幅スペクトルにおいて 1 Hz に成分がある」「(A-2) sin 波なので 1 Hz 成分の位相は $-\frac{\pi}{2}$である」

(B) 納得できないこと：「(B-1) 振幅スペクトルにおいて 31 Hz に成分がある」「(B-2) 1 Hz 成分の振幅値が 16 となっている（本来は 1 であるべき）」

が観測できます。さらに，「(B-1') 31 Hz 成分の位相は $\frac{\pi}{2}$である」も観測できます。これは，本文を読み終えた後では「納得できること」になっていることを願っています。

(3)【超難問：言葉として理解するのはタイヘンでしょうが，プログラムの書き換えはさほど困難ではないと思います。】検討対象とする波を sin 波に替えた場合について，(c) の「曲がった解釈」ができるか確認しなさい。

【答】オイラーの定理 $e^{j (2 \pi f t -\pi/2)} = \cos (2 \pi f t -\pi/2) + j \sin(2 \pi f t -\pi/2)$, $e^{j (2 \pi f t + \pi/2)} = \cos (2 \pi f t +\pi/2) + j \sin(2 \pi f t + \pi/2)$を利用して，実部と虚部を分けて描きます。実際には$e^{j (2 \pi f t -\pi/2)} = \sin (2 \pi f t ) - j \cos(2 \pi f t)$, $e^{j (2 \pi f t + \pi/2)} = - \sin (2 \pi f t) + j \cos(2 \pi f t + \pi/2)$ として，実部と虚部を描きます。具体的には以下のコードになります（書き直した部分だけ，# *** コメント *** を残しました。）。

```python
f1 = 1.0
f2 = 31.0
A  = 0.5        # 振幅は 1/2 です。

# ---- まず，実部を描きます ----
plt.figure()
fs_a = 1000.0
t_a = np.arange(0, 1, 1/fs_a)
p1t_a =  A * np.sin(2.0 * np.pi * f1 * t_a)  # ***** sin波の標本値を配列 f1_a (f_1(t) の疑似アナログ波形) に納めます。*****
p2t_a = -A * np.sin(2.0 * np.pi * f2 * t_a)  # ***** -sin波の標本値を配列 f2_a (f_2(t) の疑似アナログ波形) に納めます。*****
p3t_a = p1t_a + p2t_a

plt.title('Real part')
plot_wave(t_a, p1t_a, hold = True, color = 'b', legend = '1 Hz') 
plot_wave(t_a, p2t_a, hold = True, color = 'r', legend = '31 Hz')
plot_wave(t_a, p3t_a, hold = True, color = 'g', legend = 'Sum') 

fs = 32.0
t = np.arange(0, 1, 1/fs)
p1t =  A * np.sin(2.0 * np.pi * f1 * t)  # ***** sin波の標本値を配列 f1t (f_1(t) の標本値)に納めます。*****
p2t = -A * np.sin(2.0 * np.pi * f2 * t)  # ***** -sin波の標本値を配列 f2t (f_2(t) の標本値)に納めます。*****
p3t = f1t + f2t

plot_wave(t, p1t, hold = True, color = 'b', marker ='o') 
plot_wave(t, p2t, hold = True, color = 'r', marker ='o')
plot_wave(t, p3t,              color = 'g', marker ='o') # hold = False で，これまでに送られたデータを描画する。

# ---- 続いて，虚部を描きます（要領は実部と全く同じですので，コメントは省きます） ----
plt.figure()
p1t_a =-A * np.cos(2.0 * np.pi * f1 * t_a)  # ***** -cos に書き換え ***** 
p2t_a = A * np.cos(2.0 * np.pi * f2 * t_a)  # ***** cos に書き換え *****
p3t_a = p1t_a + p2t_a

plt.title('Imaginary part')
plot_wave(t_a, p1t_a, hold = True, color = 'b', legend = '1 Hz') 
plot_wave(t_a, p2t_a, hold = True, color = 'r', legend = '31 Hz')
plot_wave(t_a, p3t_a, hold = True, color = 'g', legend = 'Sum') 

p1t =-A * np.cos(2.0 * np.pi * f1 * t)    # ***** -cos に書き換え *****
p2t = A * np.cos(2.0 * np.pi * f2 * t)    # ***** cos に書き換え *****
p3t = p1t + p2t

plot_wave(t, p1t, hold = True, color = 'b', marker ='o') 
plot_wave(t, p2t, hold = True, color = 'r', marker ='o')
plot_wave(t, p3t,              color = 'g', marker ='o') 
```

これにより描かれた図面も，ゴチャゴチャした波形ですので，心を鎮めて見ていきます。

**実部の図面について：**  1 Hz の波（青線）を標本化した「青い●」は，31 Hz の波（赤線）を【絶妙に】標本化した「赤い●」と完全に一致していますので，影に隠れてしまい見えません。両者を重ねた「緑の●」は， 1 Hz と 31 Hz の正弦波を重ねた緑線の合成波を標本化していますが，「緑色の●」だけをみると見事に周波数が 1 Hz で振幅が 1 の sin波になっています（絶妙な標本化がなされています）。

**虚部の図面について：**  1 Hz の波（青線）を標本化した「青い●」と，31 Hz の波（赤線）を【絶妙に】標本化した「赤い●」とは，完全に正負が逆転しています。両者を重ねた「緑の●」は， 1 Hz と 31 Hz の余弦波を重ねた緑線の合成波を標本化していますが，「緑色の●」だけをみると見事に 0 が並んでいることが分かります（絶妙な標本化がなされています）。つまり，虚部が完全にキャンセルされているのです。

## 4.3 窓関数による波形の変化とスペクトルの変化

(1) 上記 (a) では，ハニング窓をオーバーラップアドすることを考えた。(b) では，継続時間の長い正弦波に対してハニング窓をかけることで切り出しを行った。それでは，(b) のように「ハニング窓関数をかけて切り出した正弦波」をオーバーラップアドすることで，(a) と同じことを確認しなさい。

【答】1500 Hz の $\cos$波を準備して，32点のハニング窓で切り出してはオーバーラップアドしてみましょう。

```python
# 継続点数 512点で周波数100 Hzの正弦波を準備
fs = 16000.0    # 標本化周波数は 16 kHz
M =  512        # 点数は 512 点
twoPi = 2.0 * np.pi

f = 1500.0      # 周波数は 1500 Hz
A = 1.0         # 振幅は 1

t = 1 / fs * np.arange(0, M)       # 1 s 間を，1 / fs [s] ごとに区切る時刻の配列
pn = A *  np.cos(twoPi * f * t)    # 標本値の配列

# 長さ32点の窓関数を準備
N = 32             # 長さ32 点の窓
wn = hanning(N)    # 窓関数の配列
shift = int(N / 2)

# 切り出し区間をシフトさせて３つの波形を切り出してから窓掛け
excerpt1 =  wn * pn[0 : N]
excerpt2 =  wn * pn[0+shift : N+shift]
excerpt3 =  wn * pn[0+2*shift : N+2*shift]

# まずは切り出した最初の部分だけを描画（正弦波の両端が潰れている）
plt.subplot(4,1,1)
plot_wave([], excerpt1, marker = 'o')

# 2つをオーバーラップアドして，全体を描画（オーバーラップ区間は正弦波は潰れていない）
sum = overlap_add(excerpt1, excerpt2)
plt.subplot(4,1,2)
plot_wave([], sum, marker = 'o')     

# 2つをオーバーラップアドして，オーバーラップしている部分だけ描画（キレイな正弦波）
plt.subplot(4,1,3)
plot_wave([], sum[shift:-shift], marker = 'o')     

# 3つをオーバーラップアドして，オーバーラップしている部分だけ描画（キレイな正弦波）
sum = overlap_add(sum, excerpt3)
plt.subplot(4,1,4)
plot_wave([], sum[shift:-shift], marker = 'o') 
```

## 4.4 忘れてはいけない位相スペクトル

(1) 【超難問】 上記 (a) で n0 ポイントだけ遅らせるように位相スペクトルを変形することを行った。そのプログラムで，n0 = 0.5 としてみなさい。波形がインパルスではなくなることに気付くはずである。何故か考えること。（ヒント：4.1節 (c) で，アナログ波形を描き，標本化された点と重ねて表示しています。それと同じことを行ってみると，「単位パルスの背景にあるアナログ波形」が見えてきます。それが見えると，0.5 ポイントを遅らせた場合の，波形の乱れ（決して乱れではない！）の理由が分かります。）

【答】まず，インプット4.17と4.19から必要な部分をコピーして，n0 = 0.5と編集したコードを以下に示します。まずは，逆FFTで得た時間波形を見てください。およそ「単位パルス」とはいえない波形となっています。

それでは，アナログ波形を重ねて見ましょう。

```python
# まずアナログ波形を描くための準備を行う
def waveHCT(t, sp, N, f0):
    '''高調波構造複合音 (HCT: Harmonic Complex tone) の波形を求める関数
    t:  時間軸データ
    sp: スペクトル
    N:  FFTの次数
    f0: 基本周波数
    '''
    twoPi = 2.0 * np.pi          # 2πを設定
    NyOrder = int(N/2)           # ナイキスト周波数に対応する次数
    sum = np.zeros(len(t))       # 足しこまれるデータを配列を用意し，0で初期化
    sum = np.real(sp[0])         # 直流成分
    sum += np.real(sp[NyOrder] * np.exp(1j * twoPi * ( NyOrder * f0) * t))  # ナイキスト周波数成分を加算
    for n in range(1, NyOrder): # 高調波成分 
        component =  sp[n]   * np.exp(1j * twoPi * ( n * f0) * t) \
                   + sp[N-n] * np.exp(1j * twoPi * (-n * f0) * t)           # 第n高調波を計算
        sum += np.real(component)                                            # 第n高調波を加算
    return(sum/N)
    
# 比較するため，疑似的なアナログ波形を描く
fs_a = 1000.0    # 疑似的にアナログ波形を描くため標本化周波数を 1000 Hz
t_a = np.arange(0, N, 1/fs_a)     # N s 間を，1 / 1000 [s] ごとに区切る時刻の配列

# 単位パルスについて波形を描く
fn_a = waveHCT(t_a, sp, N, f0=1.0/N)            # 疑似アナログ波形の標本値の配列
plot_wave(t_a, fn_a, color='r', hold = True)   # 疑似アナログデータのみ転送
plot_wave(t, unit_pulse, marker = 'o')          # ディジタルデータを送り描画

# 0.5ポイントだけ時間推移させた単位パルスについて波形を描く
fn_a = waveHCT(t_a, sp_modified, N, f0=1.0/N)   # 疑似アナログ波形の標本値の配列
plot_wave(t_a, fn_a, color='r', hold = True)   # 疑似アナログデータのみ転送
plot_wave(t, unit_pulse_modified, marker = 'o') # ディジタルデータを送り描画
```

「単位パルスの背景にあるアナログ波形」が見えたでしょう。ここでも【絶妙なサンプリング】がなされています（サンプルされるのは，波形が時間軸と交わる点）。「0.5ポイントだけ時間推移させた単位インパルス」も全く同様であり，サンプルされた点が乱れているわけではない（背景にあるアナログ時間波形が時間推移しているだけである）ことが分かります。

(3) 上記 (b) の処理を「全ての成分を sin の位相で重ねる」ように変更して実行すること。音色の変化は，cos の位相で揃えた場合に比べ，いかがか？

【答】 以下にインプット4.20と4.21について必要な変更を加えたコードを示します。

```python
# インプット 4.20
fs, wave_data = scipy.io.wavfile.read ('sample/voice[aiueo]fs16kHz.wav')
print('Sampling frequency =', fs, '[Hz]')
sampling_interval = 1.0 / fs
times = np.arange ( len ( wave_data )) * sampling_interval
plt.title("Output 4.20")
plot_wave ( times, wave_data )

audio = Audio(wave_data, rate = fs)
#audio

# インプット 4.21
buf_length = 256   # 区間の長さ
n_loop = np.floor(len(wave_data) / buf_length)  
# 最後の区間長に満たない部分は（無音区間なので）処理対象にしない

start = 0                       # 切り出し区間の開始点を初期化
result = np.array([])           # 処理結果を納める配列を初期化 
for i in range(0, int(n_loop)):
    buf = wave_data[start : start + buf_length] # start からバッファ長だけ切り出し
    sp = np.fft.fft(buf)                        # FFT して
    phi = np.array([- np.pi/2 for k in range(0, int(buf_length / 2) + 1 )])    # 正の周波数領域について計算
    phi = np.append(phi, -phi[ -2 : 0 :-1]) # 負の周波数領域では，偏角が正負逆転

    sp = np.abs(sp)* np.exp(1j * phi)          # 位相回転させて
    buf = np.real(np.fft.ifft(sp))             # 逆 FFT して 
    result = overlap_add(result, buf, 0)       # 結果をつなげる
    start += buf_length                        # 次の切り出し区間の先頭に移動

times = np.arange ( len ( result )) * sampling_interval
plt.title("Output 4.21")
plot_wave ( times , result )

audio = Audio(result, rate = fs)
audio
```

波形としては，cos波の位相で揃えた場合が「単極パルス」であったのに対して，sin波で揃えると「双極パルス」的になります。しかし，音色としては cos波で揃えた場合に比べて，ほとんど変わりません（パルス的という波形の特徴は音色としても特徴的ではありますが...）。音色知覚の妙です。
